```python
import requests
import pandas as pd
```

```python
url="https://prep2024.iecm.mx/prep2024/api_avance_entidad.php?type=1"
```

```python
r=requests.get(url)
r.status_code
```

```python
data=r.json()
data
```

```python
distrito=[]
votos_morena=[]
votos_coalicion=[]
votos_mc=[]
lista_nominal=[]
votos_nulos=[]
votos_cand_no_reg=[]

#Extraer datos por distrito
for i in range(len(data["votos_del_jg"])):
    distrito.append(data["votos_del_jg"][i]["id_distrito"])
    votos_coalicion.append(data["votos_del_jg"][i]["total_votos_cc1"])
    votos_morena.append(data["votos_del_jg"][i]["total_votos_cc5"])
    votos_mc.append(data["votos_del_jg"][i]["votos_part_6"])
    lista_nominal.append(data["votos_del_jg"][i]["lista_nominal"])
    votos_nulos.append(data["votos_del_jg"][i]["votos_nulos"])
    votos_cand_no_reg.append(data["votos_del_jg"][i]["votos_cand_no_reg"])
    
```

```python
#Ponerlo en dataframe
df=pd.DataFrame({"distrito":distrito,"pvem_pt_morena":votos_morena,"pan_pri_prd":votos_coalicion,"mc":votos_mc,"lista_nominal":lista_nominal,"votos_nulos":votos_nulos,"votos_cand_no_reg":votos_cand_no_reg})
df
```
